```python
from random import seed
from random import randint
```

```python
seed(1)
x, y = list(), list()
for i in range(100):
	in_pattern = [randint(1,100) for _ in range(2)]
	out_pattern = sum(in_pattern)
	print(in_pattern, out_pattern)
	x.append(in_pattern)
	y.append(out_pattern)
```
